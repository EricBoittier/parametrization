```python
def plot_D(test_path, skip=5):
    test_data = open(test_path).read()
    x = [float(_.split()[0]) for _ in test_data.split("\n")[:-1]]
    y = [float(_.split()[1]) for _ in test_data.split("\n")[:-1]]
    plt.scatter(x,y)

    coef = np.polyfit(x[skip:],y[skip:],1)
    fit = [x_*coef[0]+coef[1] for x_ in x]

    plt.plot(x,fit)
    plt.title("D = {:.2f} ".format(coef[0]) + "$\mathrm{\AA}^{2}$/ps")
    plt.xlabel("Time [ps]")
    plt.ylabel("MSD [$\mathrm{\AA}^{2}$]")
    print(coef)
```
