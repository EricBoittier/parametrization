```python
import arch
import arch.bootstrap as bs
import scipy.stats as stats

def boot_strap(x,y,func=np.mean):

    stat_bs = bs.StationaryBootstrap(1000, x,y=y)
    means = []

    for i in stat_bs.bootstrap(600):
        bs_x = i[0][0]
        bs_y = i[1]["y"]
        means.append(func(bs_y))

    return means

def get_se(x,y):
    means = boot_strap(x,y)
    # se = stats.sem(means)
    se = np.std(means)
    return se
```

```python
K_se = get_se(t_,K_)
v_se = get_se(t_,v_)
v2_se = get_se(t_,v_**2)
U_se = get_se(t_, U_/202)

U = ufloat(np.mean(U_/202), U_se) * ureg("kcal/mol")
U.__dict__
float(U._magnitude.nominal_value)
float(U._magnitude.std_dev)
```

```python

vol = ufloat(np.mean(v_), v_se) * ureg("angstrom^3") 
vol2 = ufloat(np.mean(v_**2), v2_se) * ureg("angstrom^6")
K = ufloat(np.mean(K_), K_se) * ureg("K")
Boltzmann_const = 1.38064852 * 10 ** (-23)  * ureg("J/K") # % J K-1


def get_kappa(Vavg, V2avg, T):
    """Calculate isothermal compressibility
    k_B in j K^-1
    :param V2avg: m^3
    :param Vavg: m^3
    :param T: K
    :return: kappa in Pa (j/m^3) converted to 1/atm
    """
    Vavg2 = Vavg * Vavg
    kappa = (V2avg - Vavg2) / (T * Boltzmann_const * Vavg)
    return kappa.to("1/atm")

kappa = get_kappa(vol, vol2, K)
kappa.__dict__
```

```python
ureg = UnitRegistry()
Avogadro_const = 6.02214129 * 10 ** 23 * ureg("1/mol")  # % mol-1

def get_density(volume, N_res, MW):
    """
    get the density of the simulation for a given volume, number of molecules and molecular weight
    :param volume: m^3
    :param N_res: number
    :param MW: g/mol
    :return: density in g/m^3
    """
    MW = MW * ureg("g/mol")

    return N_res * MW / (Avogadro_const * volume)

```
